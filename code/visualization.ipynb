```python
import plotly.express as px
import pandas as pd
```

```python
fig = px.violin(phenotypic, y = "SITE_ID", x = "AGE_AT_SCAN",
                labels = {"SITE_ID": "Site",
                        "AGE_AT_SCAN": "Age"},
                color = "SEX",
                orientation="h"
               )
fig.update_traces(width=2)
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.colors import n_colors

fig = go.Figure()

fig.add_trace(go.Violin(y = phenotypic["SITE_ID"],
                        x = phenotypic["AGE_AT_SCAN"],
                        #side="positive",
                        line_color = "light blue",
                        orientation = "h"
                       )
             )

fig.update_traces(meanline_visible = False,
                  width = 1.25,
                  points = False,
                  scalemode = "width")
fig.update_layout(violingap = 0, violinmode = "overlay",
                  title = "Age distributions at different research sites",
                  xaxis_title = "Age in Years",
                  yaxis_title = "Site",
                  height = 800,
                  yaxis = {"categoryorder": "mean ascending"}
                 )
fig.show()
```

```python
sites = list(phenotypic_repl.SITE_ID.unique())

phenotypic_repl["AGE_AT_SCAN"][phenotypic_repl["SITE_ID"] == sites[0]]

import plotly.graph_objects as go
from plotly.colors import n_colors

fig = go.Figure()

colors = n_colors('rgb(5, 200, 200)', 'rgb(200, 10, 10)', 17, colortype='rgb')

for site, color in zip(sites, colors):
    fig.add_trace(go.Violin(y = phenotypic_repl["SITE_ID"][phenotypic_repl["SITE_ID"] == site],
                        x = phenotypic_repl["AGE_AT_SCAN"][phenotypic_repl["SITE_ID"] == site],
                        #side="positive",
                        line_color = color,
                        orientation = "h",
                        marker_size = 2,
                       )
             )

fig.update_traces(meanline_visible = True,
                  points = "all",
                  scalemode = "width")
fig.update_layout(violingap = 0, violinmode = "overlay",
                  title = "Age distributions at different research sites",
                  xaxis_title = "Age in Years",
                  yaxis_title = "Site",
                  height = 800,
                  yaxis = dict(categoryorder = "mean ascending",
                               
                              )
                 )
fig.show()
```
